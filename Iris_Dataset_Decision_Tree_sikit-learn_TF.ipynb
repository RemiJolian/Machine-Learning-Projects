### A basic example in decision tree, using sikit-learn and TF



#### In this example we use Iris dataset

# Iris Dataset Overview

## Origin
- Introduced by British statistician and biologist Ronald Fisher in 1936

## Content
150 samples from three species of Iris flowers:
1. Iris setosa
2. Iris virginica
3. Iris versicolor

## Features
For each sample, four features are measured:
1. Sepal length (in cm)
2. Sepal width (in cm)
3. Petal length (in cm)
4. Petal width (in cm)

## Structure
- Total samples: 150 (50 from each species)
- Number of features: 4 (numeric)
- Number of classes: 3 (the three Iris species)

This format provides a clearer, more structured presentation of the Iris dataset information, making it easier to read and understand at a glance.

```python
from sklearn.datasets import load_iris
iris = load_iris()
X, y = iris.data, iris.target
print(y)
print(f"len(y) is :{len(y)}")
print(X)
print(f"len(y) is :{len(X)}")
#print(X[33])
```

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree # try export_graphviz
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Load the iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the decision tree classifier
clf = DecisionTreeClassifier(max_depth=3, random_state=42) # try max_depth = 4 or... 8
clf.fit(X_train, y_train)

# Evaluate the model
accuracy = clf.score(X_test, y_test)
#accuracy = clf.score(X_train, y_train) ???

print(f"Accuracy: {accuracy:.2f}")

# Visualize the decision tree
plt.figure(figsize=(20,10))
plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True)
plt.show()

# Make a prediction
sample = X_test[27].reshape(1, -1) # try X_test[0 or ...]
prediction = clf.predict(sample)
#print(prediction)
print(f"Prediction for sample: {iris.target_names[prediction[0]]}")
```


This example does the following:

1. We import the necessary libraries, including the iris dataset and the DecisionTreeClassifier from scikit-learn.
2. We load the iris dataset and split it into training and testing sets.
3. We create a DecisionTreeClassifier with a maximum depth of 3 to prevent overfitting.
4. We train the classifier on the training data.
5. We evaluate the model's accuracy on the test data.
6. We visualize the decision tree using plot_tree().
7. Finally, we make a prediction on a single sample from the test set.

This example demonstrates how to create, train, evaluate, and use a decision tree for a simple classification task. The visualization helps you understand how the tree makes decisions based on the features of the iris flowers.

When you run this code, you'll see the accuracy of the model printed, a visualization of the decision tree, and a prediction for a single sample. This provides a good starting point for understanding how decision trees work in practice.

#### TensorFlow

#### While scikit-learn is a popular choice for implementing decision trees, it's not the only option. TensorFlow can also be used to implement decision trees, although it's more commonly used for deep learning models. Let me provide you with an example of how you can use TensorFlow for decision trees.

##### Below, because of version incompatibility between TF and decision tree, maybe errors occur. So we can ran it in Google Colab. (Iris_Dataset_TF_approach.ipynb in my Google Drive)

```python
import tensorflow as tf
import tensorflow_decision_forests as tfdf
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load the iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Convert to TensorFlow datasets
train_ds = tf.data.Dataset.from_tensor_slices((X_train, y_train)).batch(32)
test_ds = tf.data.Dataset.from_tensor_slices((X_test, y_test)).batch(32)

# The batch size determines how many data samples are processed together in each training iteration.
# A batch size of 32 means the model will update its weights after looking at 32 samples.

# Create and train the model
model = tfdf.keras.RandomForestModel()
model.fit(train_ds)

# Compile the model with the accuracy metric
model.compile(metrics=["accuracy"])

# Evaluate the model 
evaluation = model.evaluate(test_ds)
print(f"Accuracy: {evaluation[1]:.2f}")

# Make a prediction
sample = tf.constant([X_test[0]])
prediction = model.predict(sample)
predicted_class = tf.argmax(prediction, axis=1).numpy()[0]
print(f"Prediction for sample: {iris.target_names[predicted_class]}")

# Display model structure
print(model.summary())
```


This example uses TensorFlow Decision Forests, which is an extension of TensorFlow for decision tree-based models. Here's what the code does:

1. We import TensorFlow and TensorFlow Decision Forests libraries.
2. We load the iris dataset and split it into training and testing sets.
3. We convert the data into TensorFlow datasets
4. We create a Random Forest model (which is an ensemble of decision trees) using TensorFlow Decision Forests.
5. We train the model on the training data.
6. We evaluate the model's accuracy on the test data.
7. We make a prediction on a single sample.
8. Finally, we display a summary of the model structure.
